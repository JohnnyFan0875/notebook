```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 
df = pd.read_csv("cvd_exon_percentage.csv")  #  Gene 
df_filtered = df.filter(regex="^percentage")

#  gene 
df_melted = df_filtered.melt(var_name="Vendor", value_name="Coverage")

# 
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x="Vendor", y="Coverage")
plt.title("Coverage Comparison Between Vendors")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
