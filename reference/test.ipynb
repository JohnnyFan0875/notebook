```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 讀取資料
df = pd.read_csv("cvd_exon_percentage.csv")  # 假設第一欄是 Gene 名稱
df_filtered = df.filter(regex="^percentage")

# 去掉 gene 名稱欄，只保留各廠商欄位
df_melted = df_filtered.melt(var_name="Vendor", value_name="Coverage")

# 畫圖
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x="Vendor", y="Coverage")
plt.title("Coverage Comparison Between Vendors")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
